```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import layers
from tensorflow.keras import models
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils, to_categorical
```

```python
lb = LabelEncoder()
```

```python
y_cat = np_utils.to_categorical(lb.fit_transform(y))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.30, random_state=1405, stratify=y)
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train.reshape(-1, X_train.shape[-1])).reshape(X_train.shape)
X_test = scaler.transform(X_test.reshape(-1, X_test.shape[-1])).reshape(X_test.shape)
```

```python
from tensorflow.keras.optimizers import Adam

custom_adam = Adam(learning_rate=0.001)
```

```python
model = models.Sequential()
model.add(layers.Masking(mask_value=0))
# model.add(layers.Bidirectional(layers.LSTM(128, return_sequences=True, activation = "tanh")))
# model.add(layers.Bidirectional(layers.LSTM(128, return_sequences=True, activation = "tanh")))
model.add(layers.LSTM(128, return_sequences=True, activation = "tanh"))
model.add(layers.LSTM(128, return_sequences=True, activation = "tanh"))
model.add(layers.Conv1D(128, 8, padding='same', activation = "tanh"))
model.add(layers.Conv1D(128, 8, padding='same', activation = "tanh"))
# model.add(layers.BatchNormalization())
model.add(layers.Dropout(0.4))
model.add(layers.MaxPooling1D(pool_size=(8)))

# model.add(layers.Conv1D(128, 8, padding='same', activation = "tanh"))
model.add(layers.Conv1D(128, 8, padding='same', activation = "tanh"))
# model.add(layers.BatchNormalization())
model.add(layers.Conv1D(126, 8, padding='same', activation = "tanh"))
model.add(layers.Dropout(0.4))
model.add(layers.Conv1D(256, 8, padding='same', activation = "tanh"))
model.add(layers.MaxPooling1D(pool_size=(8)))
model.add(layers.Flatten())
model.add(layers.Dense(100, activation='relu'))
# model.add(layers.Bidirectional(layers.LSTM(10, return_sequences=True, activation = "tanh")))
# model.add(layers.TimeDistributed(layers.Dense(8, activation='softmax')))
model.add(layers.Dense(8, activation='softmax'))
model.compile(loss="categorical_crossentropy", optimizer=custom_adam, metrics=["accuracy"])
```

```python
es = EarlyStopping(patience=15, restore_best_weights=True)
history = model.fit(X_train, y_train, validation_split=0.3, epochs=1000, batch_size=256, callbacks=[es])
```
